```python
import matplotlib.pyplot as plt

x=np.arange(0,3*np.pi,0.1)
print(x)
y=np.sin(x)

```

```python
def plotcreator(lower,upper,interval):
  '''creates a sin plot like how we do in maths,depends
  and depends upon the arange function,for without it this wont work well'''
  x=np.arange(lower,upper,interval)
  y=np.sin(x)
  plt.plot(x,y)
  plt.show()
```

```python
intervals=[0.1,0.2,0.5,1.0,2.0,5.0]
for i in intervals:
  plotcreator(-5,5,i)
```

```python
def scatterplotcreator(lower,upper,interval):
  '''creates a sin plot like how we do in maths,depends
  and depends upon the arange function,for without it this wont work well'''
  x=np.arange(lower,upper,interval)
  y=np.sin(x)
  plt.scatter(x,y)
  plt.show()
```

```python
intervals=[0.1,0.2,0.5,1.0,2.0,5.0]
for i in intervals:
  scatterplotcreator(-5,5,i)
```
